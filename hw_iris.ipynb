```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score

# Загружаем датасет Iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Оставляем только два признака: sepal length и sepal width
X = df[['sepal length (cm)', 'sepal width (cm)']]
y = iris.target

# Разделяем данные на обучающую и тестовую выборки (75% на обучение, 25% на тест)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

print("Данные загружены. Размер обучающей выборки:", X_train.shape)
X.head()
```

### 2. Построение модели LDA
Обучаем модель линейного дискриминантного анализа и визуализируем предсказания.

```python
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)

# Делаем предсказания
y_pred = lda.predict(X_test)

# Оцениваем точность (accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(f"Точность модели LDA (Accuracy): {accuracy:.2f}")

# Визуализация предсказаний LDA и центров классов
plt.figure(figsize=(10, 6))

# Рисуем точки тестовой выборки, раскрашенные по предсказанным классам
plt.scatter(X_test['sepal length (cm)'], X_test['sepal width (cm)'], c=y_pred, cmap='viridis', s=50, alpha=0.8, label='Предсказания')

# Рисуем центры классов (центроиды), найденные LDA
plt.scatter(lda.means_[:, 0], lda.means_[:, 1], c='red', s=200, marker='*', label='Центры классов')

plt.title(f'LDA: Предсказания на тестовой выборке (Accuracy: {accuracy:.2f})')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### 3. Кластеризация K-Means
Используем метод локтя для поиска оптимального `k`, а затем строим кластеризацию.

```python
# 3. Кластеризация K-Means

print("--- Подбор оптимального числа кластеров (Метод локтя) ---")

inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Визуализация метода локтя
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, marker='o')
plt.title('Метод локтя для определения оптимального k')
plt.xlabel('Количество кластеров')
plt.ylabel('Inertia (Сумма квадратов расстояний)')
plt.xticks(k_range)
plt.grid(True)
plt.show()

# Из графика "локтя" берем k=3
optimal_k = 3
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters = kmeans_final.fit_predict(X)

plt.figure(figsize=(10, 6))
plt.scatter(X['sepal length (cm)'], X['sepal width (cm)'], c=clusters, cmap='viridis', s=50, alpha=0.8)
plt.scatter(kmeans_final.cluster_centers_[:, 0], kmeans_final.cluster_centers_[:, 1], c='red', s=200, marker='*', label='Центроиды')
plt.title(f'K-Means кластеризация (k={optimal_k})')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
